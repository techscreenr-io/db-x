```python
class CleanData:
    def __init__(self, path_file):
        self.path_file = path_file
        self.data = pd.read_csv(path_file)
        self.data.fillna('',inplace=True)


    def show_data(self):
        return self.data
    
    def rename_columns(self):
        self.data.rename(columns={'Name of Business':'business_name',
                                            'Services Offered':'services',
                                            'Location':'location',
                                            'Business Contact':'phone_number',
                                            'Business Email':'email',
                                            'Social Media Handle':'twitter_handle',
                                            'Business website ':'website',
                                            'Whatsapp Link':'whatsapp_link'}, inplace=True)
        
    def clean(self):
            # twitter
        self.data['twitter_handle']=(self.data['twitter_handle'].str.replace('/','',regex=True)
                        .str.replace(r'(?<=\S)\s+(?=\S)', ', ', regex=True)
                        .str.replace('@', '',regex=True)
                        .str.replace(r'https:wa\.mec254794835416', '', regex=True)
                        .str.replace(r'https:twitter\.com','',regex=True)
                        .str.replace(r',,','',regex=True)
                        .str.replace(r'\([^)]*\)', '',regex=True)
                        )
    
            # Website
        self.data['website']=(self.data['website'].str.replace('coming soon','',regex=True)
                 .str.replace('Coming soon','',regex=True)
                 .str.replace('in progress','',regex=True)
                 .str.replace('Comming Soon','',regex=True)
                 .str.replace('Coming Soon','',regex=True)
                 .str.replace(r'https://wa\.me/.*', '', regex=True)
                 )
        
            # Email
        self.data['email']=self.data['email'].str.replace('/',', ',regex=True)

            # whatsapp_link
        self.data['whatsapp_link']=self.data['whatsapp_link'].str.replace(r'https://x\.com.*', '', regex=True)

            #location 
        self.data['location']=(self.data['location'].str.replace(r'\([^)]*\)','', regex=True)
                               .str.replace('/', ',', regex=True)
                               .str.replace('delivery','')
                               .str.replace('roles','')
                            #    .str.replace(' ','')
                               .str.replace(r'.*Country.*', 'Nationwide', regex=True,case=False)
                               .str.replace(r'.*Nation.*','Nationwide',regex=True,case=False)
                               .str.replace('freelancer','Remote',case=False)
                               .str.replace('naationwide','Nationwide')
                               .str.replace(r'.*Remote.*','Remote',regex=True,case=False)
                               .str.replace(r'.*Nairobi.*', 'Nairobi', case=False,regex=True)
                               .str.replace(r'.*Global.*','Global',case=False,regex=True)
                               .str.replace(r'.*Worldwide.*','Global',case=False,regex=True)
                               .str.replace('.*Chuka.*','Tharaka Nithi',regex=True,case=False)
                               .str.replace(r'.*Parklands.*','Nairobi',case=False,regex=True)
                            #    .str.replace()
                               )
        

             # Phone_number
        self.data['phone_number']=(self.data['phone_number'].str.replace('/',',',regex=True)
                      .str.replace(r'[^\d,]','',regex=True)
                      .str.replace(' ','')
                      .str.replace(r'^254','+254',regex=True)
                      .str.replace(r'^7','+2547',regex=True)
                      .str.replace(r'^07','+2547',regex=True)
                      .str.replace(r'^1','+2541',regex=True)
                      .str.replace(r'^01','+2541',regex=True)

                      .str.replace(',07',', +2547',regex=True)
                      .str.replace(',7',', +2547',regex=True)
                      .str.replace(',254',', +254')
                      .str.replace(',1',', +2541',regex=True)
                      .str.replace(',01',', +2541',regex=True)
                      
                      )
        

            # Load to Elasticsearch
        
        

        

    




data_cleaner = CleanData('~/Downloads/KOT-PROJECT/KOT Business DB - Sheet1.csv')


# # show data
# Rename columns
data_cleaner.rename_columns()

# clean data
data_cleaner.clean()

# show clean data


# show cleaed data
# data_cleaner.show_data().head()
```
